## Oceanbolt Python SDK - Fleet Speed of Ballasting Supramaxes in Indian Ocean Last 30 Days

Example of Oceanbolt's Python SDK to answer the question - "What was the speed of ballasting Supramaxes over the Indian Ocean in the last 30 days?" 


```python
# Import the relevant libraries
# In order to get a timeseries of fleet speed for ballasting Supramaxes in the Indian Ocean, we use the FleetSpeedTimeseries client (https://python-sdk.oceanbolt.com/tonnage_v3/fleet_speed.html)

from oceanbolt.sdk.client import APIClient
from oceanbolt.sdk.data.tonnage import FleetSpeedTimeseries
from datetime import date
```

```python
# Create the base API client using your token. Tokens can be created in the Oceanbolt App (app.oceanbolt.com)
base_client = APIClient("<token>")
```

```python
# Connect to the relevant Oceanbolt data endpoints using the base client object, ie: FleetSpeedTimeseries
speed_client = FleetSpeedTimeseries(base_client)
```

```python
# We apply filters to our query to get the timeseries of the fleet speed for ballasting Supramax vessels in the Indian Ocean
# A complete list of available filters can be found at https://python-sdk.oceanbolt.com/tonnage_v3/zone_count.html#arguments

df = speed_client.get(
   segment=['supramax'],
    laden_status=["ballast"],
    zone_id=[22, 13, 97, 96, 8], #Zone Ids can be obtained from the zone selector map in the Oceanbolt App or from the entities/zones endpoint.
    last_n_days= 30, #We select trailing 30 days.  
)

#Print 2 first rows of table
df.head(2) 
```

```python
#Let's plot the timeseries
df.plot(x='date', figsize=(10,10), title='Fleet Speed of Ballasting Supramaxes in Indian Ocean - Last 30 days');
```
