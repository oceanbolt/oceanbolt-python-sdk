## Using Oceanbolt to do Custom Polygon Counts

In this notebook we will go through how you can use the Oceanbolt SDK
to the count number of vessels within a user-defined / custom poygon.

```python
from oceanbolt.sdk.client import APIClient
from oceanbolt.sdk.data.tonnage import CustomPolygonTimeseries
from datetime import date
```

```python
base_client = APIClient("<token>")
```

We will need coordinates for the polygon to be able to populate the request body.
The gif below shows how to create polygons in your browers on https://geojson.io/ and get the required GeoJSON string.
<img src="custom_polygons.gif" width="500" align="center">

We use the copied string to populate our request body in the cell below.

```python
df = CustomPolygonTimeseries(base_client).get(
    geojson="{\"type\":\"Polygon\",\"coordinates\":[[[33.28857421875,32.91648534731439],[30.146484374999996,32.65787573695528],[31.025390625,29.36302703778376],[32.93701171875,27.235094607795503],[34.8486328125,27.410785702577023],[33.28857421875,32.91648534731439]]]}", #Polygon coordinates corresponding to polygon that we drew. 
    start_date=date(2021,1,1),
    end_date=date(2021,3,25),
    # You can specify additional parameters, e.g. laden status or segment.
)
```

```python
# Let us plot the results
df.plot(x='date', figsize=(10,5))
```
